### Code to convert our csv into sql insert into query

```python
import csv
openFile = open('student_dataset.csv', 'r')
csvFile = csv.reader(openFile)
header = next(csvFile)

headers = [h.replace(' ','') for h in header]
headers = [h.replace('.','_') for h in headers]

insert = 'INSERT INTO csv_to_sql (' + ", ".join(headers) + ") VALUES "

for row in csvFile:
    values = [f"'{v.strip()}'" if v and not v.isdigit() else v.strip() if v else 'NULL' for v in row]
    insert_row = insert + "(" + ", ".join(values) + ");"
    print(insert_row)
openFile.close()
```

### Code flow:
- Open the CSV file using open() in read mode and store it in the openFile variable. <br>
- Create a csv.reader object using the csv module and pass the openFile variable to it. Store the resulting object   in csvFile.<br>
- Use next() to read the first row of the CSV file, which contains the column headers. Store it in header.<br>
- Clean up the header row by removing spaces using the replace() method  & replace dots -->underscores.<br>
- Create the initial INSERT statement by using join() to combine the elements of headers with commas between them.   Store the statement in insert.<br>
- Loop through each row of csvFile using a for loop.<br>
    - Clean up the row data by checking each value to see if it is a string or a number using isdigit(). 
    - If it is a string, wrap it in single quotes using strip(). If it is a number, leave it as is. If it is empty,       replace it with the NULL keyword.<br>
    - Use join() to combine the elements of values with commas between them. Add the resulting string to the end of       the INSERT statement to create a complete INSERT INTO query.
    - Print insert_row to generated SQL query.
    - Close the file using close() method on openFile.

### **Subquery** <br>
In SQL, a subquery is a query that is nested within another query, also known as the outer query. A subquery can be used to return a set of rows that will be used in the outer query as a condition to filter the results. A subquery can be used in the SELECT, FROM, WHERE, and HAVING clauses of a query.
<br>

*code flow:*<br>
- This SQL query selects all the information from the csv_to_sql table where the salary of the employee is equal to   the maximum salary found in the salary column of the same employees table.

- It does this by using a subquery that finds the maximum salary in the salary column of the csv_to_sql table, and   then using that result to filter the main query and return only the rows where the salary matches the maximum       salary.

```python
# value_counts() in python.

query = "select location, count(*) as count_of_candidate from csv_to_sql GROUP BY location;"
city = pd.read_sql_query(query, conn)
city
```

### flow:<br> 
Here's how the query works step-by-step:
- The SELECT statement selects two columns: "location" and "COUNT(*) as count_of_candidate".
- The "COUNT(*)" function counts the number of rows for each group of location.
- The "GROUP BY" clause groups the rows by city name.
- The result set shows the unique city names along with the count of employees for each location.

```python
query = "select location, count(*) as count_of_candidate from csv_to_sql GROUP BY location ORDER BY count_of_candidate DEsc;"
city = pd.read_sql_query(query, conn)
city
```

### Flow of DB

A high-level overview of how the information flows in a typical database system:

- A user sends a query to the database management system (DBMS) via an interface, such as a command-line prompt, a   web form, or an API.

- The DBMS receives the query and checks its syntax to ensure it is valid.

- The DBMS then looks up the requested data in its data dictionary to determine the structure of the data and the     tables that contain it.

- The DBMS then uses an optimizer to determine the most efficient way to execute the query, taking into account the   size and complexity of the database, the available hardware resources, and any indexes or constraints on the       data.

- The DBMS executes the query and returns the results to the user, either directly or via an application layer.
